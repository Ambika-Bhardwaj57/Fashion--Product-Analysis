# Data Visualization

```python
data.columns
```

```python
sns.lineplot(data['Price'])
plt.show()
```

```python
sns.displot(data['Rating'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Product Count by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Price', bins=10)
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Product Price Distribution')
plt.show()
```

```python
plt.figure(figsize=(40, 30))
sns.boxplot(data=data, x='Price', y='Rating')
plt.xlabel('Price')
plt.ylabel('Rating')
plt.title('Price vs Rating')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Category', y='Price')
plt.xlabel('Category')
plt.ylabel('Price')
plt.title('Product Price by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Set up the figure and axes
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot 1: Bar plot of categories
axs[0, 0].set_title('Category Distribution')
sns.countplot(x='Category', data=data, ax=axs[0, 0])
axs[0, 0].tick_params(axis='x', rotation=45)

# Plot 2: Histogram of prices
axs[0, 1].set_title('Price Distribution')
sns.histplot(data=data, x='Price', kde=True, ax=axs[0, 1])

# Plot 3: Box plot of ratings
axs[1, 0].set_title('Rating Distribution')
sns.boxplot(data=data, x='Rating', ax=axs[1, 0])

# Plot 4: Bar plot of colors
axs[1, 1].set_title('Color Distribution')
sns.countplot(x='Color', data=data, ax=axs[1, 1])
axs[1, 1].tick_params(axis='x', rotation=45)

# Adjust the layout
plt.tight_layout()

# Display the plots
plt.show()
```

```python
columns = ['Price', 'Rating']

# Compute the correlation matrix
corr_matrix = data[columns].corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

# Customize the plot
plt.title('Correlation Heatmap')
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(8, 6))
data['Brand'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Brands')
plt.ylabel('')
plt.show()
```

 ✨✨✨  Nike is the best Brand

```python
# Let's see the brand with the highest average rating
# And the maximum rating for each brand
fig, ax = plt.subplots(figsize=(8,4), dpi=200, facecolor='#1a1a1a')
ax.set_facecolor('#1a1a1a')
(data
 .groupby('Brand')
 .Rating
 .agg(['max', 'mean'])
 .reset_index()
 .plot.bar(x='Brand', y=['max', 'mean'],
           color=['#FFFF00', '#7CFC00'],
           rot=360, ax=ax, width=.8)
)
fig.suptitle('Rating: max and mean', fontsize=14, fontweight='bold', color='white')
ax.spines[['left','right','bottom','top']].set_visible(False)
ax.tick_params(axis='both', colors='white');
ax.legend('');
```

```python
data_Brand = data
data_Brand['Total Sales'] = data_Brand.iloc[:, 1:].sum(axis=1, numeric_only=True)

# Sort the data by total sales in descending order
data_Brand_sorted = data_Brand.sort_values('Total Sales', ascending=False)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(data_Brand_sorted['Product Name'], data_Brand_sorted['Total Sales'])

plt.xlabel('Brand')
plt.ylabel('Price')
plt.title('Top Product')
plt.xticks(rotation=45)
plt.show()
```
